```python
feature_with_na=[feature for feature in train.columns if train[feature].isnull().sum()>=1]
```

```python
train.isnull().sum()/train.shape[0]*100
```

```python
# missing values percentage
for feature in feature_with_na:
    print(feature,np.round(train[feature].isnull().mean()*100,5), '% of missing values')
    
```

```python
# let'splot some amezing diagrams

for feature in feature_with_na:
        data=train.copy()
        # let's make a variables that indicate 1 if the observation was missing
        data[feature]=np.where(data[feature].isnull(),1,0)
        
        # let's calculate the mean saleprice where the information is missing of pre
        data.groupby(feature)['SalePrice'].median().plot.bar()
        plt.title(feature)
        plt.show()
```

## Numerical Variable

```python
# list of numerical variable
numerical_feature=train.dtypes[train.dtypes!='O'].keys()
numerical_feature
```

```python
numerical_feature=[feature for feature in train.columns if train[feature].dtypes !="O"]
numerical_feature
```

```python
print(len(numerical_feature))
```

```python
train[numerical_feature].head(5)
```

```python
train.info()
```

```python
year_feature=[feature for feature in numerical_feature if "Yr" in feature or "Year" in feature]
```

```python
year_feature
```

```python
for feature in year_feature:
    print(feature,train[feature].unique())
    print('=='*20)
```

```python
train.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('year sold')
plt.ylabel('median house price')
plt.title('house prive vs year sold')
```

```python
# we will compare the difference b/w all years feature with salesprice
for feature in year_feature:
    if feature !='YrSold':
        train[feature]=train['YrSold']-train[feature]
        
        plt.scatter(train[feature],train['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

# obserbations
whitin all the yearwise information,it is clearly visible that the new houses are more costilier that the older one. looks like a pareto or powerlaw distribution


```python
# numerical variable -2 types
# 1 continous variavle and Discrete variavle
discrete_feature=[feature for feature in numerical_feature if len(train[feature].unique())<=25]
```

```python
print(len(discrete_feature))
```

```python
train[discrete_feature].head(5)
```

```python
## lets find the relation between discrete and sales price

data=train.copy()
```

```python
for feature in discrete_feature:
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('saleprice')
    plt.title(feature)
    plt.show()
```

```python
## continuous variable
continuous_feature=[feature for feature in numerical_feature if feature not in discrete_feature+year_feature]
len(continuous_feature)
```

```python
## lets analyze the continous values by creating histogram to understand the distribution
data=train.copy()
```

```python
for feature in continuous_feature:
    data.groupby(feature)['SalePrice'].median().plot.hist()
    plt.xlabel(feature)
    plt.ylabel('saleprice')
    plt.title(feature)
    plt.show()
    
```

```python
for feature in continuous_feature:
    data.groupby(feature)['SalePrice'].median().plot.hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('saleprice')
    plt.title(feature)
    plt.show()
```

# EDA part 2

```python
# we will using logarithmic transformation
data=train.copy()
for feature in continuous_feature:
    if 0 in data[feature].unique():
        pass 
    else: 
        data[feature]=np.log(data[feature])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```

# Outliers

```python
# Outliers
data=train.copy()
for feature in continuous_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
